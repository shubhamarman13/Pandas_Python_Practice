```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Generating Indian student names
student_names = [
    'Aarav', 'Vivaan', 'Aditya', 'Vihaan', 'Arjun',
    'Sai', 'Reyansh', 'Ayaan', 'Anaya', 'Naira',
    'Saanvi', 'Ishaan', 'Riya', 'Kaira', 'Krishna',
    'Dhruv', 'Lakshmi', 'Rohan', 'Anika', 'Aditi',
    'Dev', 'Anushka', 'Shiv', 'Tanvi', 'Meera',
    'Radhika', 'Kabir', 'Ava', 'Nia', 'Yash',
    'Pooja', 'Sneha', 'Neha', 'Ritika', 'Siddharth',
    'Tara', 'Karan', 'Ritika', 'Aryan', 'Anvi',
    'Shreya', 'Mohan', 'Rahul', 'Sonia', 'Aishwarya',
    'Kavya', 'Aarushi', 'Rani', 'Rohit', 'Pankaj'
]

# Generating random marks for 5 subjects
marks = np.random.randint(0, 101, size=(50, 5))

# Defining subject names
subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']

# Creating the DataFrame
df = pd.DataFrame(marks, columns=subjects)
df['Name'] = student_names  # Adding the names
df['Total'] = df[subjects].sum(axis=1)  # Calculate the total marks

# Determining pass/fail based on the average marks
df['Result'] = np.where(df['Total'] / 5 >= 40, 'Pass', 'Fail')

# Reordering columns to ensure 'Name' is the first column
df = df[['Name'] + subjects + ['Total', 'Result']]

print(df)
```
